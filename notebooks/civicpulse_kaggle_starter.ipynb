# CivicPulse Fairness Audit Starter

This notebook mirrors the backend audit flow for Kaggle publication.

```python
import pandas as pd

df = pd.read_csv('../data/sample_housing_assistance.csv')
df.head()
```

```python
def selection_rate(frame, group_col, score_col='model_score', threshold=0.5):
    return frame.assign(pred=(frame[score_col] >= threshold).astype(int)).groupby(group_col)['pred'].mean()

rates = selection_rate(df, 'district')
rates
```

```python
disparate_impact = rates.min() / rates.max()
demographic_parity_difference = rates.max() - rates.min()
disparate_impact, demographic_parity_difference
```
